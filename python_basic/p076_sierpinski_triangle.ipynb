## Sierpinski Tiangle - Matplotlib

```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
def transformation_1(x,y):
    x_next = 0.5*x
    y_next = 0.5*y
    return x_next,y_next

def transformation_2(x,y):
    x_next = (0.5*x)+0.5
    y_next = (0.5*y)+0.5
    return x_next,y_next

def transformation_3(x,y):
    x_next = (0.5*x)+1
    y_next = 0.5*y
    return x_next,y_next
```

```python
x_values = [0]
y_values = [0]
transformation=[transformation_1,transformation_2,transformation_3]
def triangle(num):
    x,y = 0,0
    for i in range(num):
        chosen = choice(transformation)
        x,y = chosen(x,y)
        x_values.append(x)
        y_values.append(y)
```

```python
triangle(1000000)
plt.rc('figure',figsize=(16,16))
plt.plot(x_values,y_values,'o')
plt.show
```
